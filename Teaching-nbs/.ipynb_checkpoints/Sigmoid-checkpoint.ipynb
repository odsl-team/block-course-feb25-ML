```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sig(x):
    return 1 / (1 + np.exp(-x))
```

```python
plt.figure(figsize=(6,6))

xx = np.linspace(-7,7)

plt.plot(xx,sig(xx),color='limegreen',label='$\sigma(z)$',lw=3)

ax = plt.gca()
for loc in ['right','top']:
    ax.spines[loc].set_visible(False)

ax.spines['bottom'].set_position(('data', 0))
# ax.spines['left'].set_position(('data', 0))

ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

ax.plot((xx[-1]), (0), ls="", marker=">", 
    ms=16, color="k", clip_on=False)

ax.plot((xx[0]), (1), ls="", marker="^", 
    ms=16, color="k", clip_on=False)

plt.xlim(xx[[0,-1]])
plt.ylim(0,1.01)


plt.xlabel('z',loc='right',fontsize=20)

plt.legend(loc='upper left',frameon=False,fontsize=20)
plt.savefig('figures/sigmoid.pdf',bbox_inches='tight')

plt.plot(xx,sig(5*xx),color='deeppink',label='$\sigma(5 z)$',lw=3)
plt.plot(xx,sig(xx/5),color='darkorange',label='$\sigma(z / 5)$',lw=3)
plt.legend(loc='upper left',frameon=False,fontsize=20)

plt.savefig('figures/sigmoids.pdf',bbox_inches='tight')
```

```python
plt.figure(figsize=(6,6))

xx = np.linspace(-7,7)

plt.plot(xx,sig(xx),color='limegreen',label='$\sigma(z)$',lw=3)
plt.plot(xx,sig(xx)*( 1- sig(xx)),'k--',label='$\sigma\'(z)$',lw=3)

ax = plt.gca()
for loc in ['right','top']:
    ax.spines[loc].set_visible(False)

ax.spines['bottom'].set_position(('data', 0))
# ax.spines['left'].set_position(('data', 0))

ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

ax.plot((xx[-1]), (0), ls="", marker=">", 
    ms=16, color="k", clip_on=False)

ax.plot((xx[0]), (1), ls="", marker="^", 
    ms=16, color="k", clip_on=False)

plt.xlim(xx[[0,-1]])
plt.ylim(0,1)


plt.xlabel('z',loc='right',fontsize=20)

plt.legend(loc='upper left',frameon=False,fontsize=20)
plt.savefig('figures/sigmoid_derivative.pdf',bbox_inches='tight')
```

```python
plt.figure(figsize=(6,6))

xx = np.linspace(-7,7,300)

plt.plot(xx,sig(xx),color='limegreen',label='$\sigma(z)$',lw=3)
plt.plot(xx,sig(5*xx),color='deeppink',label='$\sigma(5 z)$',lw=3)
plt.plot(xx,sig(xx/5),color='darkorange',label='$\sigma(z / 5)$',lw=3)

plt.plot(xx,sig(xx)*(1-sig(xx)),color='limegreen',
         label='$\sigma\'(z)$',ls='--',lw=3)
plt.plot(xx,5*sig(5*xx)*(1-sig(5*xx)),color='deeppink',
         label='$5 \sigma\'(5 z)$',ls='--',lw=3)
plt.plot(xx,sig(xx/5)*(1- sig(xx/5))/5,color='darkorange',
         label=r"$\frac{1}{5} \sigma'(z / 5)$",ls='--',lw=3)

ax = plt.gca()
for loc in ['right','top']:
    ax.spines[loc].set_visible(False)

ax.spines['bottom'].set_position(('data', 0))
# ax.spines['left'].set_position(('data', 0))

ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

ymax=1.3
ax.plot((xx[-1]), (0), ls="", marker=">", 
    ms=16, color="k", clip_on=False)

ax.plot((xx[0]), (ymax), ls="", marker="^", 
    ms=16, color="k", clip_on=False)

plt.xlim(xx[[0,-1]])
plt.ylim(0,ymax)

plt.xlabel('z',loc='right',fontsize=20)

plt.legend(loc='upper left',frameon=False,fontsize=16)
plt.savefig('figures/sigmoid_derivative_3.pdf',bbox_inches='tight')
```

```python
plt.figure(figsize=(6,6))

xx = np.linspace(-7,7)

plt.plot(xx,sig(xx),color='k',label='$\sigma(z)$',lw=3)

plt.plot(xx,sig(xx) + sig(-xx),'C0--',label='$\sigma(z)(1-\sigma(z))$',lw=3)
plt.plot(xx,sig(xx)*( 1 - sig(xx)),'C2--',label='$\sigma(z)(1-\sigma(z))$',lw=3)
plt.plot(xx,sig(xx)*( 1 + sig(-xx)),'C3--',label='$\sigma(z)(1+\sigma(-z))$',lw=3)
plt.plot(xx,sig(xx)*( 1 - sig(-xx)),'C4--',label='$\sigma(z)(1-\sigma(-z))$',lw=3)

ax = plt.gca()
for loc in ['right','top']:
    ax.spines[loc].set_visible(False)

ax.spines['bottom'].set_position(('data', 0))
# ax.spines['left'].set_position(('data', 0))

ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

ax.plot((xx[-1]), (0), ls="", marker=">", 
    ms=16, color="k", clip_on=False)

ax.plot((xx[0]), (1), ls="", marker="^", 
    ms=16, color="k", clip_on=False)

plt.xlim(xx[[0,-1]])
plt.ylim(0,1)


plt.xlabel('z',loc='right',fontsize=20)

plt.legend(loc='upper left',frameon=False,fontsize=14)
plt.savefig('figures/sigmoid_derivative_multipleChoice.pdf',bbox_inches='tight')
```

**ROC curve:**

Let's keep working with the cats + dogs examples and colors, and just try to make a histogram bounded b/w (0,1) for these two classes

```python
?np.random.exponential
```

```python
Nex = 100000

s0 = 4*np.random.random(Nex) - 6+np.random.exponential(1,Nex)
s1 = 5*np.random.random(Nex) + 4 - 2*np.random.exponential(1,Nex)
```

```python
?np.random.uniform
```

```python
h0,e,_ = plt.hist(np.hstack([sig(s0),
                             np.random.uniform(size=Nex//4)]),
                  20,(0,1),color='red',label='cat',
                  alpha=.5,density=True)
h1,_,_ = plt.hist(np.hstack([sig(s1),
                             np.random.uniform(size=Nex//4)]),
                  20,(0,1),color='dodgerblue',label='dog',
                  alpha=.5,density=True)

ax = plt.gca()
for loc in ['right','top']:
    ax.spines[loc].set_visible(False)

ax.spines['bottom'].set_position(('data', 0))
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

xmax = 1.05
_,ymax = plt.ylim()

ax.plot((xmax), (0), ls="", marker=">", ms=16, color="k", clip_on=False)
ax.plot((0), (ymax), ls="", marker="^", ms=16, color="k", clip_on=False)

plt.xlim(0,xmax)
plt.ylim(0,ymax)

plt.xlabel(r'$f(x) = \sigma(\theta^T x)$',fontsize=20)
plt.ylabel('Entries',fontsize=20)

plt.legend(loc='upper left',frameon=False,fontsize=20)
plt.savefig('figures/disc_hists.pdf',bbox_inches='tight')

plt.show()
```

```python
eff_B = np.add.accumulate(h0[::-1])/np.sum(h0)
```

```python
eff_S = np.add.accumulate(h1[::-1])/np.sum(h1)
```

```python
xx = .5 * (e[1:]+e[:-1])
```

```python
np.sum(xx)
```

```python
auc = np.sum(xx*eff_S)/np.sum(xx)
```

```python
plt.plot(eff_B, eff_S,color='dodgerblue',lw=2,label=f'classifier: AUC = {auc:.2f}')

ax = plt.gca()
for loc in ['right','top']:
    ax.spines[loc].set_visible(False)

ax.spines['bottom'].set_position(('data', 0))
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

xmax = 1.05
_,ymax = plt.ylim()

ax.plot((xmax), (0), ls="", marker=">", ms=16, color="k", clip_on=False)
ax.plot((0), (ymax), ls="", marker="^", ms=16, color="k", clip_on=False)

plt.xlim(0,xmax)
plt.ylim(0,ymax)

plt.xlabel(r'False Positive Rate',fontsize=20)
plt.ylabel(r'True Positive Rate',fontsize=20)

plt.plot([0,1],[0,1],color='grey',ls='dotted',lw=2,label='random: AUC = 0.5')
plt.legend(fontsize=16,frameon=False)

plt.savefig('figures/roc.pdf',bbox_inches='tight')
```
