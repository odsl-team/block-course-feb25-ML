# Regulartization

**Goal:** Plot the L1 and L2 losses

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(-2,2)
```

```python
plt.xlabel('w',fontsize=18)
plt.ylabel('Loss',fontsize=18)
plt.plot(xx, xx**2,'dodgerblue',label='L2 loss',lw=4)

plt.xlim(-2,2)
plt.legend(fontsize=18)
plt.savefig('figures/L2.png')

# And now add the L1 loss
plt.plot(xx,np.abs(xx),'crimson',label='L1 loss',lw=4)

plt.xlim(-2,2)
plt.legend(fontsize=18)
plt.savefig('figures/L2-L1.png')
```
