**Goal:** Build up some theory for optimizers and optimization algorithms.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
```

```python
# In python, arrays are passed by reference, so if modifying an array, want
# to modeify a copy, not the original object
from copy import copy
```

```python
'''
Functions used in Lili Jiang's blog post
https://towardsdatascience.com/a-visual-explanation-of-gradient-descent-methods-momentum-adagrad-rmsprop-adam-f898b102325c
https://github.com/lilipads/gradient_descent_viz/blob/master/src/gradient_descent.cpp
'''

def local_minimum(x,z):
    z = copy(z)
    z *= 1.4
    return -2 * np.exp(-5*((x - 1)**2 + z**2)) - 6. * np.exp(-5*((x + 1)**2 + z**2)) + x**2 + z**2

def global_minimum(x,z):
    return x * x + z * z

def saddle_point(x,z):
    return np.sin(x) + z * z

def ecliptic_bowl(x,z):
    x = copy(x)
    z = copy(z)
    x /= 2.
    
    z /= 2.
    return -exp(-(x * x + 5 * z * z)) + x * x + 0.5 * z * z

def hills(x,z):
    z = copy(z)
    z *= 1.4;
    return  2 * np.exp(-((x - 1) * (x - 1) + z * z) / .2) + \
            6. * np.exp(-((x + 1) * (x + 1) + z * z) / .2) - \
            2 * np.exp(-((x - 1) * (x  - 1) + (z + 1) * (z + 1)) / .2) + \
            x * x + z * z

def plateau(x,z):
    x = copy(x)
    z = copy(z)
    
    x *= 10
    z *= 10
    r = np.sqrt(z * z + x * x) + 0.01
    return -np.sin(r) / r + 0.01 * r * r
```

I think it would be fun to vizualize the benefit of momentum in a 1d plot

$$\mathcal{L}(\theta) = -2 \exp \left(-5 (x - 1)^2  \right) - 6 \exp \left(- 5 (x + 1)^2 \right)+ x^2$$

$$\frac{\partial \mathcal{L}}{\partial \theta} = 20 (x-1) \exp \left(-5 (x - 1)^2  \right) + 60 (x+1) \exp \left(- 5 (x + 1)^2 \right)+ 2x$$

```python
def dlocal_minimum(x,z):
    assert z==0
    return 20 * (x-1) * np.exp(-5 * (x-1)**2) + 60 * (x+1) * np.exp(-5 * (x+1)**2) + 2*x 
```

```python
x = np.linspace(-2,2)
z = 0
l = local_minimum(x,z)
dl = dlocal_minimum(x,z)

plt.plot(x,l,lw=4,color='grey',label=r'$L(\theta)$')
plt.plot(x,dl,lw=4,color='k',ls='--',label=r'$dL/d\theta$')
plt.plot([-2,2],[0]*2,'r')
plt.xlim(-2,2)
plt.xlabel(r'$\theta$',fontsize=20)
plt.ylabel(r'$\mathcal{L}(\theta)$',fontsize=28)

# Turn off the spines
ax = plt.gca()
ax.spines[['right', 'top']].set_visible(False)
ax.spines[:].set_linewidth(4)
plt.legend(fontsize=18)
```

```python
x = np.linspace(-2,2)
z = 0
l = local_minimum(x,z)

plt.plot(x,l,lw=4,color='grey')
plt.xlim(-2,2)
plt.xlabel(r'$\theta$',fontsize=20)
plt.ylabel(r'$\mathcal{L}(\theta)$',fontsize=28)

# Turn off the spines
ax = plt.gca()
ax.spines[['right', 'top']].set_visible(False)
ax.spines[:].set_linewidth(4)

# Set an initial point
theta_i = 1.5
plt.scatter([theta_i],[local_minimum(theta_i,z)],color='turquoise',zorder=5,s=121)

'''
SGD
'''
alpha = 1e-2

iters = 40
sgd = pd.DataFrame(0,index=range(iters+1),columns=['theta_t','Loss','deriv_loss'])
sgd.loc[0] = [theta_i,local_minimum(theta_i,z),dlocal_minimum(theta_i,z)] # initial

cmap = mpl.colormaps['GnBu']
for i,c in zip(range(iters),cmap(np.linspace(.5, 1, iters))):

    deriv_i = dlocal_minimum(theta_i,z)
    theta_i += (- alpha * deriv_i)

    loss_i = local_minimum(theta_i,z)
    sgd.loc[i+1] = [theta_i,loss_i,deriv_i] # initial

    plt.scatter([theta_i],[loss_i],color=c,zorder=5,s=121,alpha=.5)

print(ax.get_ylim())
plt.show()
```

Note, unless you tune the learning rate, $\alpha$, to fly out of this local minimum, SGD wonm't be able to find the true global minima of the function, which is at $\theta = -1$.

```python
#### x = np.linspace(-2,2)
z = 0
l = local_minimum(x,z)

plt.plot(x,l,lw=4,color='grey')
plt.xlim(-2,2)
plt.xlabel(r'$\theta$',fontsize=20)
plt.ylabel(r'$\mathcal{L}(\theta)$',fontsize=28)

# Turn off the spines
ax = plt.gca()
ax.spines[['right', 'top']].set_visible(False)
ax.spines[:].set_linewidth(4)


iters = 5


# (Optionally) overlay SGD
cmap = mpl.colormaps['GnBu']
colors = list(cmap(np.linspace(0.5, 1, iters)))
plt.scatter(sgd.theta_t[:iters+1], sgd.Loss[:iters+1], color=['turquoise']+colors,zorder=5,s=121,alpha=.5)

# Set an initial point
theta_i = 1.5
plt.scatter([theta_i],[local_minimum(theta_i,z)],color='purple',zorder=5,s=121)

'''
Momentum
'''
alpha =  1e-2 #from sgd case
delta = .9

iters = 40
mom = pd.DataFrame(0,index=range(iters+1),columns=['theta_t','Loss','deriv_loss'])
mom.loc[0] = [theta_i,local_minimum(theta_i,z),dlocal_minimum(theta_i,z)] # initial

deriv_prev = 0

cmap = mpl.colormaps['PuRd']
for i,c in zip(range(iters),cmap(np.linspace(0.5, 1, iters))[::-1]):

    deriv_i = dlocal_minimum(theta_i,z)
    theta_i -= alpha * (deriv_i + delta * deriv_prev)

    # Remember our history
    deriv_prev = deriv_i
    
    loss_i = local_minimum(theta_i,z)
    mom.loc[i+1] = [theta_i,loss_i,deriv_i] # initial

    plt.scatter([theta_i],[loss_i],color=c,zorder=5,s=121,alpha=.5)

plt.ylim(-5.456985605644304, 4.436215044651645)
plt.show()
```

**Note**, with momentum, we optimize _more quickly_ I can also adjust the parameters to fly _out_ of the bowl, but I can't get it to stably go to the $2^{nd}$ minima if I initialize to $\theta \leftarrow 1.5$.
